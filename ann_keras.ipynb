```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



#configuration
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
_d=pd.read_csv('./sample_data/train.csv')
_d.drop(['Id'],axis=1,inplace=True)
y=_d['SalePrice']
x=_d.drop(['SalePrice'],axis=1)
```

```python
x.head()
```

```python
cat_columns=x.select_dtypes('object').columns
int_columns=x.select_dtypes(['int64','float64']).columns
```

```python
#sklearn
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
cat_pipeline=Pipeline([
                       ('impute',SimpleImputer(strategy='constant',fill_value='missing')),
                       ('encoder',OneHotEncoder(drop='first'))
])

num_pipeline=Pipeline([
                       ('impute',SimpleImputer()),
                      ('scaling',StandardScaler())
])

column_transformer=ColumnTransformer([
                   ('cat',cat_pipeline,cat_columns),
                   ('numerical',num_pipeline,int_columns)
],
remainder='passthrough')
_transformed_data=column_transformer.fit_transform(x)
```

```python
x=_transformed_data.toarray()
```

```python
_transformed_data.toarray().shape[1]
```

```python
#keras
import keras
from keras.models import Sequential
from keras.layers import Dense,PReLU,LeakyReLU,ELU,Dropout

```

```python
_transformed_data.toarray().shape[1]
```

```python
model=Sequential()
model.add(Dense('10',activation='relu',input_shape=(261,)))
model.add(Dense('10',activation='relu'))
model.compile(optimizer='Adamax',loss='mean_absolute_error')
model.fit(x,y,validation_split=0.33,batch_size=10,epochs=100)
```
